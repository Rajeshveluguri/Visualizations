```python
from adjustText import adjust_text

import numpy as np
from matplotlib import pyplot as plt
```

<H3> Basic Example </H3>

Without Adjust Text

```python
np.random.seed(33)

x, y = np.random.random((2, 20))
```

```python
plt.scatter(x, y)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]
```

With Adjust Text - 

```python
plt.scatter(x, y)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]

adjust_text(texts)
```

With Seaborn

```python
import seaborn as sns

sns.set_style('darkgrid')

penguins = sns.load_dataset('penguins').dropna().sample(25, random_state=23)
```

```python
penguins.head()
```

```python
x = penguins.bill_length_mm
y = penguins.bill_depth_mm
labels = penguins.species

sns.scatterplot(x=x, y=y, hue=labels, s=70, legend=False)

texts = [plt.text(x_pos, y_pos, f'{l}') for (x_pos, y_pos, l) in zip(x, y, labels)];
```

With Adjust Text

```python
x = penguins.bill_length_mm
y = penguins.bill_depth_mm
labels = penguins.species

sns.scatterplot(x=x, y=y, hue=labels, s=70, legend=False)

texts = [plt.text(x_pos, y_pos, f'{l}') for (x_pos, y_pos, l) in zip(x, y, labels)]

adjust_text(texts);
```

```python
plt.figure(figsize=(2, 3))

x_thumb = [2.5, 3.25, 3]
y_thumb = [2.5, 2, 3]
labels = ['Value 2', 'Value 3', 'Value 1']

plt.scatter(x_thumb, y_thumb, s=100, color='crimson')
plt.xlim(0, 6)
plt.ylim(0, 8)

texts = [plt.text(x_thumb[i], y_thumb[i], f'{labels[i]}') for i in range(len(x_thumb))]

#plt.savefig('adjustText_figure_before.png', dpi=200)
```

```python
plt.figure(figsize=(2, 3))

x_thumb = [2.5, 3.25, 3]
y_thumb = [2.5, 2, 3]
labels = ['Value 2', 'Value 3', 'Value 1']

plt.scatter(x_thumb, y_thumb, s=100, color='crimson')
plt.xlim(0, 6)
plt.ylim(0, 8)

texts = [plt.text(x_thumb[i], y_thumb[i], f'{labels[i]}') for i in range(len(x_thumb))]

adjust_text(texts, expand_points=(1.3, 1.3))
```
