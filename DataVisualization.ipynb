Scatter plots are not ideal when you have plenty of data points. Scatter plots get too dense to interpret when you have many data points.
Instead, use Hexbin plots.

They bin the chart into hexagons and assign a color intensity based on the number of points in that region

<h1> Hexbin Plot </h1>

```python
import numpy as np
import seaborn as sns
sns.set_theme(style="ticks")

rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x=x, y=y, kind="hex", color="#4CB391")
```

### Pie Plot
- Pie plots are useful for showing the relative proportions of the different data points in a dataset.

```python
# declaring data
data = [144, 45, 240, 75, 39]
keys = ['Group A', 'Group B', 'Group C', 'Group D', 'Group E']
  
explode = [0, 0.1, 0, 0, 0]
palette_color = sns.color_palette('bright')
  
plt.pie(data, labels=keys, colors=palette_color,
        explode=explode, autopct='%.0f%%')

plt.show()
```

```python
# Define your data
data = [25, 35, 30, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

# Define Seaborn color palette to use
explode = [0, 0.1, 0, 0]
colors = sns.color_palette('bright')[0:4]


# Create pie chart
plt.pie(data, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.show()
```

```python
# Without Exploding

# Define your data
data = [25, 35, 30, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

# Define Seaborn color palette to use
colors = sns.color_palette('bright')[0:4]


# Create pie chart
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```

### Subplots

```python
import numpy as np
  
# making subplots objects
fig, ax = plt.subplots(2, 2)
  
# create data
x = np.linspace(0, 10, 1000)
  
# draw graph
ax[0, 1].plot(x, np.sin(x), 'r-.')
ax[0, 0].plot(x, np.cos(x), 'g--')
ax[1, 0].plot(x, np.tan(x), 'y-')
ax[1, 1].plot(x, np.sinc(x), 'c.-')
  
plt.show()
```

### Relational Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

# Create a relational plot
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker")

# Display the plot
plt.show()
```

#### Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a count plot
sns.countplot(data=tips, x="day")

# Display the plot
plt.show()
```

#### Violin Plot

A violin plot (violinplot) combines a box plot with a kernel density plot, showing the distribution, probability density, and central tendencies of the data. It provides a detailed view of the datas distribution, highlighting variations and multimodalities.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot
sns.violinplot(data=tips, x="day", y="total_bill")

# Display the plot
plt.show()
```

#### Strip Plot
A strip plot (stripplot) displays individual data points for one or more categorical variables, often overlaid on a box or violin plot. It shows the distribution and concentration of data points, highlighting any potential outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a strip plot
sns.stripplot(data=tips, x="day", y="total_bill")

# Display the plot
plt.show()
```

#### Swarm Plot
A swarm plot displays individual data points for one or more categorical variables, similar to a strip plot, but adjusts points to avoid overlap. It provides a clear view of the distribution and density of the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a swarm plot
sns.swarmplot(data=tips, x="day", y="total_bill")

# Display the plot
plt.show()
```

#### Kernel Density Plot
A kernel density estimate plot (kdeplot) visualizes the probability density function of a continuous variable by smoothing the histogram with a kernel function. It provides a smooth representation of the datas distribution, allowing for better understanding of its shape and characteristics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a KDE plot
sns.kdeplot(data=tips, x="total_bill")

# Display the plot
plt.show()
```

#### Distribution Plot

A distribution plot (displot) is a versatile seaborn function that allows for visualization of univariate distributions. It can combine histograms, kernel density estimates, and rug plots to provide insights into the distribution of a single variable.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a distribution plot
sns.displot(data=tips, x="total_bill", kind="kde")

# Display the plot
plt.show()
```

#### Empirical Cumulative Distribution Function Plot (ecdfplot)
An empirical cumulative distribution function plot (ecdfplot) displays the cumulative distribution of a continuous variable based on the observed data points. It shows how the data is spread across different percentiles and can be useful for comparing distributions or assessing goodness-of-fit.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create an ECDF plot
sns.ecdfplot(data=tips, x="total_bill")

# Display the plot
plt.show()
```

#### Rug Plot
A rug plot (rugplot) is a simple plot that displays individual data points along a single axis, usually the x-axis, as small lines or ticks. It provides a visual representation of the data distribution and density, often used in combination with other types of plots like histograms or KDE plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a rug plot
sns.rugplot(data=tips, x="total_bill")

# Display the plot
plt.show()
```

#### Cluster Map

A cluster map (clustermap) is a heatmap that organizes rows and columns of a dataset based on their similarity, often using hierarchical clustering. Its useful for identifying patterns and relationships in complex datasets by grouping similar rows and columns together.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example flights dataset
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Create a cluster map
sns.clustermap(flights_pivot, cmap="viridis", standard_scale=1)

# Display the plot
plt.show()
```

#### Grid Plot

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
    
iris = sns.load_dataset("iris") 

g = sns.PairGrid(data = iris)
g.map(plt.scatter) #All the grids plot scatter plot
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
